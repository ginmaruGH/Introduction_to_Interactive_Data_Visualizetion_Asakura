```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import plotly.express as px
from IPython.display import display

pd.set_option("display.max_columns", None)
```

# 01 Plotly Express

## 1-1 Plotly Expressとは

### 1-1-1 インタラクティブな可視化事例

```python
gapminder = px.data.gapminder()
display(gapminder)
print(type(gapminder))
```

```python
gapminder.plot.scatter(x="gdpPercap", y="lifeExp", logx=True, xlim=[100, 1e6])
```

```python
# マウスホバーで要素の値が確認できるようにする
px.scatter(
    gapminder, x="gdpPercap", y="lifeExp", log_x=True, hover_name="country"
).show()
```

```python
# 人口の情報を大きさで表現
# 点の大きさの上限を指定
# 大陸ごとに色分け
# 凡例をクリックすることで描画対象の表示・非表示を切り替える
px.scatter(
    gapminder,
    x="gdpPercap",
    y="lifeExp",
    log_x=True,
    hover_name="country",
    size="pop",
    size_max=40,
    color="continent"
).show()
```

```python
# 大陸ごとに散布図を分割

facet_fig = px.scatter(
    gapminder,
    x="gdpPercap",
    y="lifeExp",
    log_x=True,
    hover_name="country",
    size="pop",
    size_max=40,
    color="continent",
    facet_col="continent",
    width=800
)
facet_fig.update_xaxes(tickfont={"size": 8})
facet_fig.show()
```

```python
# 大陸ごとに散布図を分割
# 年ごとに分割

animation_fig = px.scatter(
     gapminder,
    x="gdpPercap",
    y="lifeExp",
    log_x=True,
    hover_name="country",
    size="pop",
    size_max=40,
    color="continent",
    facet_col="continent",
    width=800,
    animation_frame="year",
)
animation_fig.update_xaxes(tickfont={"size": 8})
animation_fig.show()
```
